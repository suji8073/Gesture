```python
import os

from natsort import natsorted
import pandas as pd
import numpy as np
```

# 1번 하나의 파일로 합치기

```python
experiment_log = [f for f in os.listdir("./log/") if "experiment" in f]
experiment_log = natsorted(experiment_log)
trajectory_log = [f for f in os.listdir("./log/") if "trajectory" in f]
trajectory_log = natsorted(trajectory_log)
```

```python
logs_list = []
for pid, f in enumerate(experiment_log, 1):
    log = pd.read_table("./log/" + f)
    log["participants"] = pid
    logs_list.append(log)

experiment_df = pd.concat(logs_list, ignore_index=True)
experiment_df = experiment_df[["participants", "Block", "Target", "Error", "Time"]]
experiment_df.head()
```

```python
logs_list = []
for pid, f in enumerate(trajectory_log, 1):
    log = pd.read_table("./log/" + f)
    log["participants"] = pid
    logs_list.append(log)

trajectory_df = pd.concat(logs_list, ignore_index=True)
trajectory_df = trajectory_df[
    ["participants", "Block", "Target", "posX", "posY", "posZ", "revZ"]
]
trajectory_df.head()
```

```python
data = pd.merge(experiment_df, trajectory_df, on=["participants", "Block", "Target"], how="outer")
display(data)
data.shape
```

# 2번 experiment_log file에 error = 0만 남기기

```python
data_zero = data[data["Error"] == 0]
display(data_zero)
data_zero.shape
```

# 4번 0.04 이상 이하 구별하기

```python
end_find = dt[['revZ']]
end_find = end_find[1:]

end_find = end_find.reset_index(drop = False)
end_find = end_find[['revZ']]

end_find.loc[1232059] = [0.0]

dt["end_find"] = dt['revZ'].subtract(end_find['revZ'], axis = 0) 
```

```python
dt["Find"] = dt["revZ"] - dt["revZ"].shift(fill_value=0)
dt["start"] = dt["Find"].apply(lambda x: 1 if abs(x) >= 0.06 else 0)

dt["endi"] = dt["end_find"].apply(lambda x: 1 if( abs(x) >=0.06) else 0)
dt.iloc[1232059,13] = 1


# data.to_csv("data.csv", index=False)
```

 # 5번 trajectory log 의 x y z 의 시작점이 0, 0, 0 이 되게 변환

```python
data["cumsum_endi"] = data["start"].cumsum()
data[["posX", "posY", "posZ"]] = (
    data[["cumsum_endi", "posX", "posY", "posZ"]]
    .groupby("cumsum_endi")
    .diff()
    .fillna(0)
)
data.drop(["cumsum_endi"], axis=1, inplace=True)
data.head()
```

```python
data = data[['participants', 'Block', 'Target', 'Time', 'posX', 'posY', 'posZ', 'endi']]
data
```

# length를 구하기 위함

```python
end_point = data["endi"] == 1
Experiment_log = data[end_point]

Experiment_log = Experiment_log.reset_index(drop = False)
Experiment_log
```

```python
Experiment_log["length"] = abs(Experiment_log["index"] - Experiment_log["index"].shift(fill_value=0))
Experiment_log = Experiment_log[['participants', 'Block', 'Target', 'Time','length', 'posX', 'posY', 'posZ']]

Experiment_log
```
